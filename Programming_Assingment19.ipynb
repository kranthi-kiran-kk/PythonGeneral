```python
# #Assignment 19
# Question1
# Create a function that takes a string and returns a string in which each character is repeated once.
# Examples
# double_char("String") ➞ "SSttrriinngg"

# double_char("Hello World!") ➞ "HHeelllloo  WWoorrlldd!!"

# double_char("1234!_ ") ➞ "11223344!!__  "


def double_char(string_in):
    string_out = ""
    for ch in string_in:
        string_out = string_out + ch*2
    return string_out

input_string = input("Enter the string ")
output_string = double_char(input_string)

print(output_string)
```

```python

# Question2
# Create a function that reverses a boolean value and returns the string "boolean expected" 
# if another variable type is given.
# Examples
# reverse(True) ➞ False

# reverse(False) ➞ True

# reverse(0) ➞ "boolean expected"

# reverse(None) ➞ "boolean expected"

def reverse(boolean):
    boolean = str(boolean)
    if boolean == "True":
        return False
    elif boolean == "False":
        return True
    else:
        return "boolean expected"

k = input(" Enter the Boolean value ")
print(f"reverse({k}) : {reverse(k)}")

```

```python

# Question3
# Create a function that returns the thickness (in meters) of a piece of paper 
# after folding it n number of times. The paper starts off with a thickness of 0.5mm.
# Examples
# num_layers(1) ➞ "0.001m"
# # Paper folded once is 1mm (equal to 0.001m)

# num_layers(4) ➞ "0.008m"
# # Paper folded 4 times is 8mm (equal to 0.008m)

# num_layers(21) ➞ "1048.576m"
# # Paper folded 21 times is 1048576mm (equal to 1048.576m)


def num_layers(num):
    import math
    return (pow(2,num)*0.5)/1000,(pow(2,num)*0.5)

folding = int(input("Enter the number of foldings"))
m_out,mm_out = num_layers(folding)

print(f"Paper folded {folding} time/times is {int(mm_out)}mm (equal to {m_out}m)")


```

```python

# Question4
# Create a function that takes a single string as argument and returns an ordered list containing the 
# indices of all capital letters in the string.
# Examples
# index_of_caps("eDaBiT") ➞ [1, 3, 5]

# index_of_caps("eQuINoX") ➞ [1, 3, 4, 6]

# index_of_caps("determine") ➞ []

# index_of_caps("STRIKE") ➞ [0, 1, 2, 3, 4, 5]

# index_of_caps("sUn") ➞ [1]

def index_of_caps(str_in):
    index = []
    for i in range(len(str_in)):
        if str_in[i].isupper():
            index.append(i)
    return index

input_str = input("Enter the string")
output = index_of_caps(input_str)

print(f'index_of_caps("{input_str}") : {output}')
```

```python

# Question5
# Using list comprehensions, create a function that finds all even numbers from 1 to the given number.
# Examples
# find_even_nums(8) ➞ [2, 4, 6, 8]

# find_even_nums(4) ➞ [2, 4]

# find_even_nums(2) ➞ [2]

def find_even_nums(num):
    evens = [x for x in range(1,num+1) if x%2 == 0]
    return evens

input_num = int(input("Enter the number "))

print(f"find_even_nums(input_num) : {find_even_nums(input_num)}")


```
