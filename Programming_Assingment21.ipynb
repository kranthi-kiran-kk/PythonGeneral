```python
# Assignment 21
# Question1
# Write a function that takes a list and a number as arguments. Add the number to the end of the list, 
# then remove the first element of the list. The function should then return the updated list.
# Examples
# next_in_line([5, 6, 7, 8, 9], 1) ➞ [6, 7, 8, 9, 1]

# next_in_line([7, 6, 3, 23, 17], 10) ➞ [6, 3, 23, 17, 10]

# next_in_line([1, 10, 20, 42 ], 6) ➞ [10, 20, 42, 6]

# next_in_line([], 6) ➞ "No list has been selected"

def next_in_line(list_in,num):
    if len(list_in) == 0:
        return "No list has been selected"
    else:
        list_in.append(num)
        del list_in[0]
        return list_in
try:
    list_input = input("Enter the list").split(",")
    list_input = [int(x) for x in list_input]
except Exception as e:
    if e:
        list_input = []
number = int(input("Enter the number"))
list_output = next_in_line(list_input,number)

print(f'next_in_line({list_input},{number}) : {list_output}')

```

```python
# Question2
# Create the function that takes a list of dictionaries and returns the sum of people's budgets.
# Examples
# get_budgets([
#   { "name": "John", "age": 21, "budget": 23000 },
#   { "name": "Steve",  "age": 32, "budget": 40000 },
#   { "name": "Martin",  "age": 16, "budget": 2700 }
# ]) ➞ 65700

# get_budgets([
#   { "name": "John",  "age": 21, "budget": 29000 },
#   { "name": "Steve",  "age": 32, "budget": 32000 },
#   { "name": "Martin",  "age": 16, "budget": 1600 }
# ]) ➞ 62600


def get_budgets(list_dict):
    list_in = []
    for ele in list_dict:
        list_in.append(ele["budget"])
    return sum(list_in)
list_dict =  [
  { "name": "John", "age": 21, "budget": 23000 },
  { "name": "Steve",  "age": 32, "budget": 40000 },
  { "name": "Martin",  "age": 16, "budget": 2700 }
]
out = get_budgets(list_dict)

print(f' get_budgets({list_dict}) is {out}')
```

```python
# Question3
# Create a function that takes a string and returns a string with its letters in alphabetical order.
# Examples
# alphabet_soup("hello") ➞ "ehllo"

# alphabet_soup("edabit") ➞ "abdeit"

# alphabet_soup("hacker") ➞ "acehkr"

# alphabet_soup("geek") ➞ "eegk"

# alphabet_soup("javascript") ➞ "aacijprstv"

def alphabet_soup(str_in):
    list_in = [str_in[i] for i in range(len(str_in))]
    list_in = sorted(list_in)
    str_out = ""
    for ele in list_in:
        str_out = str_out + ele
    return str_out

input_string = input("Enter the string ")
#print(input_string[0])
print(f'alphabet_soup({input_string}) is {alphabet_soup(input_string)}')
    
```

```python
# Question4
# Suppose that you invest $10,000 for 10 years at an interest rate of 6% compounded monthly. 
# What will be the value of your investment at the end of the 10 year period?
# Create a function that accepts the principal p, the term in years t, the interest rate r, 
# and the number of compounding periods per year n. 
# The function returns the value at the end of term rounded to the nearest cent.
# For the example above:
# compound_interest(10000, 10, 0.06, 12) ➞ 18193.97
# Note that the interest rate is given as a decimal and n=12 because with monthly compounding there are 12 periods per year. 
#Compounding can also be done annually, quarterly, weekly, or daily.
# Examples
# compound_interest(100, 1, 0.05, 1) ➞ 105.0

# compound_interest(3500, 15, 0.1, 4) ➞ 15399.26

# compound_interest(100000, 20, 0.15, 365) ➞ 2007316.26

def compound_interest(p,t,r,n):
    import math
    interest = p * pow((1 + (r/n)),(n*t))
    return (round(interest,2))

p = float(input(" Enter the principal amount "))
t = int(input(" Enter the number of yaers "))
r = float(input("Enter the rate of interst in decimal values(Ex : 0.06 for 6%)" ))
n = int(input("Enter the compunding periods per year "))

print(f'compound_interest({p}, {t}, {r}, {n}) is {compound_interest(p, t, r, n)}')



```

```python
# Question5
# Write a function that takes a list of elements and returns only the integers.
# Examples
# return_only_integer([9, 2, "space", "car", "lion", 16]) ➞ [9, 2, 16]

# return_only_integer(["hello", 81, "basketball", 123, "fox"]) ➞ [81, 123]

# return_only_integer([10, "121", 56, 20, "car", 3, "lion"]) ➞ [10, 56, 20, 3]

# return_only_integer(["String",  True,  3.3,  1]) ➞ [1]



def return_only_integer(list_in):
    list_out= []
    for ele in list_in:
        if type(ele) == int:
            list_out.append(ele)
    return list_out

print(f'return_only_integer([9, 2, "space", "car", "lion", 16]) : {return_only_integer([9, 2, "space", "car", "lion", 16])}')
print(f'return_only_integer(["hello", 81, "basketball", 123, "fox"]) : {return_only_integer(["hello", 81, "basketball", 123, "fox"])}')
print(f'return_only_integer([10, "121", 56, 20, "car", 3, "lion"]) : {return_only_integer([10, "121", 56, 20, "car", 3, "lion"])}')
print(f'return_only_integer(["String",  True,  3.3,  1]) : {return_only_integer(["String",  True,  3.3,  1])}')
```
