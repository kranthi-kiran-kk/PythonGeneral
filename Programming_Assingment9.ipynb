```python
# 1.	Write a Python program to check if the given number is a Disarium Number?

number = eval(input("Enter the number you want to check : "))

def disarium(n):
    import math
    temp = n
    k = []
    m = 0
    while temp > 0:
        s = temp % 10 
        temp = temp // 10
        k.append(s)
        
    for i in range(1,len(k)+1):
        m = m + math.pow(k[-i],i)
      
    if m == n:
        print(f"The given number {n} is a Disarium Number ")
        
    else:
        print(f"The given number {n} is not a Disarium Number ")
    

disarium(number)
```

```python
# 2.	Write a Python program to print all disarium numbers between 1 to 100?

def disarium(n):
    import math
    temp = n
    k = []
    m = 0
    while temp > 0:
        s = temp % 10 
        temp = temp // 10
        k.append(s)
        
    for i in range(1,len(k)+1):
        m = m + math.pow(k[-i],i)
      
    if m == n:
        return True
        
    else:
        return False
        
for i in range(1,101):
    if disarium(i) == True:
        print(i, end = " ")
    
```

```python
# 3.	Write a Python program to check if the given number is Happy Number?

number = eval(input("Enter the number you want to check : "))

def happy_num(n):
    import math
    temp = n
    k = []
    m = 0
    while temp > 0:
        s = temp % 10 
        temp = temp // 10
        k.append(s)
        
    for i in range(1,len(k)+1):
        m = m + math.pow(k[-i],2)
        
    if m == 1:
        print(f"The given number is a Happy Number ")
        
    elif m > 9:
        happy_num(m)
    else :
         print(f"The given number is not a Happy Number ")
    
happy_num(number)

```

```python
# 4.	Write a Python program to print all happy numbers between 1 and 100?

def isHappyNumber(num):    
    rem = sum = 0  
    while(num > 0):    
        rem = num%10    
        sum = sum + (rem*rem)   
        num = num//10   
    return sum     
print("List of happy numbers between 1 and 100: ")
k = []  
for i in range(1, 101):    
    result = i;    
    while(result != 1 and result != 4):    
        result = isHappyNumber(result)    
    if(result == 1):
        k.append(i)
print(k)
```
