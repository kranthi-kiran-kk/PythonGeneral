```python
# Assignment 17

# Question1. Create a function that takes three arguments a, b, c and returns the sum of the numbers that are 
# evenly divided by c from the range a, b inclusive.
# Examples
# evenly_divisible(1, 10, 20) ➞ 0
# # No number between 1 and 10 can be evenly divided by 20.

# evenly_divisible(1, 10, 2) ➞ 30
# # 2 + 4 + 6 + 8 + 10 = 30

# evenly_divisible(1, 10, 3) ➞ 18
# # 3 + 6 + 9 = 18


def evenly_divisible(a,b,c):
    list1 = []
    for i in range(a,b+1):
        if i%c == 0:
            list1.append(i)
    return sum(list1)

input_num = (input("Enter the values of a,b,c ")).split(",")
#print(input_num)
#print(input_num[0],input_num[1],input_num[2])
sum_out = evenly_divisible(int(input_num[0]),int(input_num[1]),int(input_num[2]))

print(f"Sum of numbers is {sum_out}")
```

```python

# Question2. Create a function that returns True if a given inequality expression is correct and False otherwise.
# Examples
# correct_signs("3 < 7 < 11") ➞ True

# correct_signs("13 > 44 > 33 > 1") ➞ False

# correct_signs("1 < 2 < 6 < 9 > 3") ➞ True

def correct_signs(input_exp):
    return eval(input_exp)

input_exp = (input("Enter the expression "))
print(correct_signs(input_exp))

```

```python
# Question3. Create a function that replaces all the vowels in a string with a specified character.
# Examples
# replace_vowels("the aardvark", "#") ➞ "th# ##rdv#rk"

# replace_vowels("minnie mouse", "?") ➞ "m?nn?? m??s?"

# replace_vowels("shakespeare", "*") ➞ "sh*k*sp**r*"


def replace_vowels(input_str,ch):
    for i in input_str:
        if i in ['a','e','i','o','u','A','E','I','O','U']:
            input_str = input_str.replace(i,ch)
    return input_str

input_string = (input("Enter the string and character")).split(", ")
output_string = replace_vowels(input_string[0],input_string[1])
print(output_string)
```

```python
# Question4. Write a function that calculates the factorial of a number recursively.
# Examples
# factorial(5) ➞ 120

# factorial(3) ➞ 6

# factorial(1) ➞ 1

# factorial(0) ➞ 1

def factorial(num):
    if num == 0:
        return 1
    else:
        return (num * factorial(num-1))
    
num = int(input(" Enter the value of num "))
fact = factorial(num)
print(f"Factorial of the given number is {fact}")
        
            
```

```python
# Question 5
# Hamming distance is the number of characters that differ between two strings.
# To illustrate:
# String1: "abcbba"
# String2: "abcbda"

# Hamming Distance: 1 - "b" vs. "d" is the only difference.
# Create a function that computes the hamming distance between two strings.
# Examples
# hamming_distance("abcde", "bcdef") ➞ 5

# hamming_distance("abcde", "abcde") ➞ 0

# hamming_distance("strong", "strung") ➞ 1

def hamming_distance(string1,string2):
    count = 0
    if len(string1) != len(string2):
        print(" Strings are of not equal length")
    else:
        for i in range(len(string1)):
            if string1[i] != string2[i]:
                count += 1
        return count
    
string1 = input("Enter the first string ")
string2 = input("Enter the second string ")

distance = hamming_distance(string1,string2)
print(f"Hamming distance between two strings in {distance}")
```
