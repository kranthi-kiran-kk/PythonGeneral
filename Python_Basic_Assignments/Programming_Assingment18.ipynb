```python
# Assignment 18
# Question 1
# Create a function that takes a list of non-negative integers and strings and return a new list without the strings.
# Examples
# filter_list([1, 2, "a", "b"]) ➞ [1, 2]

# filter_list([1, "a", "b", 0, 15]) ➞ [1, 0, 15]

# filter_list([1, 2, "aasf", "1", "123", 123]) ➞ [1, 2, 123]


def filter_list(list_in):
    list_out= []
    for ele in list_in:
        if type(ele) != str:
            list_out.append(ele)
    return list_out

list_out1 = filter_list([1, 2, "a", "b"])
print(list_out1)
```

```python

# Question 2
# The "Reverser" takes a string as input and returns that string in reverse order, with the opposite case.
# Examples
# reverse("Hello World") ➞ "DLROw OLLEh"

# reverse("ReVeRsE") ➞ "eSrEvEr"

# reverse("Radar") ➞ "RADAr"

def reverse(str_in):
    ch_out = ""
    for ch in str_in:
        if ch.islower():
            ch_out = ch_out + ch.upper()
        elif ch.isupper():
            ch_out = ch_out + ch.lower()
        else:
            ch_out = ch_out + ch
    return ch_out[::-1]

input_str = str(input("Enter the string"))
output_str = print(reverse(input_str))
            
        

```

```python
# Question 4
# Write a function that calculates the factorial of a number recursively.
# Examples
# factorial(5) ➞ 120

# factorial(3) ➞ 6

# factorial(1) ➞ 1

# factorial(0) ➞ 1

def factorial(num):
    if num == 0:
        return 1
    else:
        return (num * factorial(num-1))
    
num = int(input(" Enter the value of num "))
fact = factorial(num)
print(f"Factorial of the given number is {fact}")
```

```python
# Question 5
# Write a function that moves all elements of one type to the end of the list.
# Examples
# move_to_end([1, 3, 2, 4, 4, 1], 1) ➞ [3, 2, 4, 4, 1, 1]
# # Move all the 1s to the end of the array.

# move_to_end([7, 8, 9, 1, 2, 3, 4], 9) ➞ [7, 8, 1, 2, 3, 4, 9]

# move_to_end(["a", "a", "a", "b"], "a") ➞ ["b", "a", "a", "a"]


def move_to_end(list_in,ele):
    list_out = []
    list_end = []
    for item in list_in:
        if item == ele:
            list_end.append(item)
        else:
            list_out.append(item)
    for item in list_end:
        list_out.append(item)
    return list_out

print(f"move_to_end([1, 3, 2, 4, 4, 1], 1) is equal to {move_to_end([1, 3, 2, 4, 4, 1], 1)}")
print(f"move_to_end([7, 8, 9, 1, 2, 3, 4], 9) is equal to {move_to_end([7, 8, 9, 1, 2, 3, 4], 9)}")
print(f'move_to_end(["a", "a", "a", "b"], "a") is equal to {move_to_end(["a", "a", "a", "b"], "a")}')
```
