```python
# Assignment 20
# Question1
# Create a function that takes a list of strings and integers, and filters out the list so that it returns a list of integers only.
# Examples
# filter_list([1, 2, 3, "a", "b", 4]) ➞ [1, 2, 3, 4]

# filter_list(["A", 0, "Edabit", 1729, "Python", "1729"]) ➞ [0, 1729]

# filter_list(["Nothing", "here"]) ➞ []



def filter_list(list_in):
    list_out = []
    for ele in list_in:
        if type(ele) == int:
            list_out.append(ele)
    return list_out

print(f'filter_list([1, 2, 3, "a", "b", 4]) : {filter_list([1, 2, 3, "a", "b", 4])}')
print(f'filter_list(["A", 0, "Edabit", 1729, "Python", "1729"]) : {filter_list(["A", 0, "Edabit", 1729, "Python", "1729"])}')
print(f'filter_list(["Nothing", "here"]) : {filter_list(["Nothing", "here"])}')
```

```python

# Question2
# Given a list of numbers, create a function which returns the list but with each element's index in the list added to itself. This means you add 0 to the number at index 0, add 1 to the number at index 1, etc...
# Examples
# add_indexes([0, 0, 0, 0, 0]) ➞ [0, 1, 2, 3, 4]

# add_indexes([1, 2, 3, 4, 5]) ➞ [1, 3, 5, 7, 9]

# add_indexes([5, 4, 3, 2, 1]) ➞ [5, 5, 5, 5, 5]



def add_indexes(list_in):
    list_out = []
    for i in range(len(list_in)):
        list_out.append(i+ int(list_in[i]))
    return list_out

input_list = input("Enter the values of list").split(",")
print(f'add_indexes([{input_list}]) : {add_indexes(input_list)}')

```

```python
# Question3
# Create a function that takes the height and radius of a cone as arguments and returns the volume of the cone rounded to the nearest hundredth. See the resources tab for the formula.
 
# Examples
# cone_volume(3, 2) ➞ 12.57

# cone_volume(15, 6) ➞ 565.49

# cone_volume(18, 0) ➞ 0


def cone_volume(h, r):
    import math
    volume = math.pi * h * pow(r, 2) / 3
    return volume


height = float(input("Enter the height of the cone "))
radius = float(input("Enter the radius of the cone "))

print(f'Volume of the cone of radius({radius}) and height({height}) is {round(cone_volume(height, radius),2)} cubic units')
```

```python
# Question4
# This Triangular Number Sequence is generated from a pattern of dots that form a triangle. The first 5 numbers of the sequence, or dots, are: 
# 1, 3, 6, 10, 15
# This means that the first triangle has just one dot, the second one has three dots, the third one has 6 dots and so on.
# Write a function that gives the number of dots with its corresponding triangle number of the sequence.
# Examples
# triangle(1) ➞ 1

# triangle(6) ➞ 21

# triangle(215) ➞ 23220


def triangle(num_in):
    num_out = []
    initial = 0
    for i in range(num_in):
        k = initial + i + 1
        num_out.append(k)
        initial = k
    return num_out[num_in - 1]


input_num = int(input("Enter the triangle number "))
print(f'triangle({input_num}) is {triangle(input_num)}')


```

```python
# Question5
# Create a function that takes a list of numbers between 1 and 10 (excluding one number) and returns the missing number.
# Examples
# missing_num([1, 2, 3, 4, 6, 7, 8, 9, 10]) ➞ 5

# missing_num([7, 2, 3, 6, 5, 9, 1, 4, 8]) ➞ 10

# missing_num([10, 5, 1, 2, 4, 6, 8, 3, 9]) ➞ 7



def missing_num(list_in):
    if len(list_in) != 9:
        return "Enter list of 9 numbers"
    return sum(list(range(1, 11))) - sum(list_in)


list_in_1 = [1, 2, 3, 4, 6, 7, 8, 9, 10]
list_in_2 = [7, 2, 3, 6, 5, 9, 1, 4, 8]
list_in_3 = [10, 5, 1, 2, 4, 6, 8, 3, 9]
print(f'missing_num({list_in_1}) : {missing_num(list_in_1)}')
print(f'missing_num({list_in_2}) : {missing_num(list_in_2)}')
print(f'missing_num({list_in_3}) : {missing_num(list_in_3)}')

```
