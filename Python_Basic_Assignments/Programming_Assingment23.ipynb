```python
# #Assignment 23
# Question 1
# Create a function that takes a number as an argument and returns True or False depending on 
# whether the number is symmetrical or not. A number is symmetrical when it is the same as its reverse.
# Examples
# is_symmetrical(7227) ➞ True

# is_symmetrical(12567) ➞ False

# is_symmetrical(44444444) ➞ True

# is_symmetrical(9939) ➞ False

# is_symmetrical(1112111) ➞ True


def is_symmetrical(num):
    if num == num[::-1]:
        return True
    else:
        return False
    
number = input("Enter the number to check")
print(f'is_symmetrical({int(number)}) : {is_symmetrical(number)}')
```

```python

# Question 2
# Given a string of numbers separated by a comma and space, return the product of the numbers.
# Examples
# multiply_nums("2, 3") ➞ 6

# multiply_nums("1, 2, 3, 4") ➞ 24

# multiply_nums("54, 75, 453, 0") ➞ 0

# multiply_nums("10, -2") ➞ -20

def multiply_nums(str_in):
    list_in = str_in.split(", ")
    list_in = [int(x) for x in list_in]
    product = 1
    for ele in list_in:
        product = product*ele
    return product

string_input = input("Enter the string of numbers seperated by comma and space")

print(f'multiply_nums("{string_input}") : {multiply_nums(string_input)}')
```

```python
# Question 3
# Create a function that squares every digit of a number.
# Examples
# square_digits(9119) ➞ 811181

# square_digits(2483) ➞ 416649

# square_digits(3212) ➞ 9414
# Notes
# The function receives an integer and must return an integer.

def square_digits(str_in):
    list_in = [str_in[x] for x in range(len(str_in))]
    list_out = [pow(int(ele),2) for ele in list_in]
    ch = ""
    for ele in list_out:
        ch = ch + str(ele)
    return ch

number = input("Enter the number ")

print(f'square_digits({number}) : {square_digits(number)}')

```

```python
# Question 4
# Create a function that sorts a list and removes all duplicate items from it.
# Examples
# setify([1, 3, 3, 5, 5]) ➞ [1, 3, 5]

# setify([4, 4, 4, 4]) ➞ [4]

# setify([5, 7, 8, 9, 10, 15]) ➞ [5, 7, 8, 9, 10, 15]

# setify([3, 3, 3, 2, 1]) ➞ [1, 2, 3]

def setify(list_in):
    return list(set(sorted(list_in)))

print(f'setify([1, 3, 3, 5, 5]) : {setify([1, 3, 3, 5, 5])}')
print(f'setify(4, 4, 4, 4]) : {setify([4, 4, 4, 4])}')
print(f'setify([5, 7, 8, 9, 10, 15]) : {setify([5, 7, 8, 9, 10, 15])}')
print(f'setify([3, 3, 3, 2, 1]) : {setify([3, 3, 3, 2, 1])}')


```

```python
# Question 5
# Create a function that returns the mean of all digits.
# Examples
# mean(42) ➞ 3

# mean(12345) ➞ 3

# mean(666) ➞ 6
# Notes
# •	The mean of all digits is the sum of digits / how many digits there are (e.g. mean of digits in 512 is (5+1+2)/3(number of digits) = 8/3=2).
# •	The mean will always be an integer.

def mean(str_in):
    list_in = [int(str_in[x]) for x in range(len(str_in))]
    mean_num = int(sum(list_in)/len(list_in))
    return mean_num

number = input("Enter the number ")

print(f'mean({number}) : {mean(number)}')
    
```
