```python
#1.	Write a Python Program to Add Two Matrices?


i = eval(input("Enter number of rows of Matrix   "))
j = eval(input("Enter number of columns of Matrix  "))
A = []
B = []
Sum = []
for k in range(1,i+1):
    C = []
    for m in range(1,j+1):
        n = eval(input(f"Enter the value of A[{k}][{m}] element  "))
        C.append(n)
        if m == j:
            A.append(C)
            
for k in range(1,i+1):
    C = []
    for m in range(1,j+1):
        n = eval(input(f"Enter the value of B[{k}][{m}] element  "))
        C.append(n)
        if m == j:
            B.append(C)       

print(f"Fisrt Matrix = {A}")
print(f"Second Matrix = {B}")

for k in range(1,i+1):
    C = []
    for m in range(1,j+1):
        n = A[k-1][m-1] + B[k-1][m-1]
        C.append(n)
        if m == j:
            Sum.append(C)    

print(f"Sum of two Matrices is : {Sum}")
```

```python
# 3.	Write a Python Program to Transpose a Matrix?

def matrix_generation():
    i = eval(input("Enter number of rows of Matrix   "))
    j = eval(input("Enter number of columns of Matrix  "))
    A = []
    for k in range(1,i+1):
        C = []
        for m in range(1,j+1):
            n = eval(input(f"Enter the value of Matrix[{k}][{m}] element  "))
            C.append(n)
            if m == j:
                A.append(C)
    return A
def matrix_zero(i,j):
    A = []
    for k in range(1,i+1):
        C = []
        for m in range(1,j+1):
            n = 0
            C.append(n)
            if m == j:
                A.append(C)
    return A


def matrix_transpose():
    A = matrix_generation()
    B = matrix_zero(len(A),len(A[0]))
    for k in range(0,i):
        C = []
        for m in range(0,j): 
            B[k][m] = A[m][k]
            C.append(B[k][m])
            if m == j:
                B.append(C)
    return B

print(f"Transpose of given Matrix {A} is {matrix_transpose()}")
```

```python
# 4.	Write a Python Program to Sort Words in Alphabetic Order?

def sort_string(S):
    W = S.split(" ")
    for i in range(len(W)):
        W[i]=W[i].lower() 
        S = sorted(W)
    print(f"New string with words sorted :") 
    print(' '.join(S))
    
    
        
String_input = str(input("Enter the string to be sorted \n"))
 

sort_string(String_input)
```

```python
# 5.	Write a Python Program to Remove Punctuation From a String?


def remove_punc(S):
    
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    remove_punct = ""
    for char in S:
        if char not in punctuations:
            remove_punct = remove_punct + char
    print(remove_punct)

String_input = str(input("Enter the string from which punctuations are to be removed \n"))

remove_punc(String_input)
```
